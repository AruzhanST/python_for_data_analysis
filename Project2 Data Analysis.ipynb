# In which countries do developers get paid the most?

```python
# Let's eliminate NaN values for 'CompTotal' column

data=df[df['CompTotal'].notnull()]
data
```

```python
well_paid_data = data.copy()

well_paid_data = well_paid_data[['Country', 'Currency', 'CompTotal', 'ConvertedCompYearly']]
well_paid_data
```

```python
well_paid_data['Currency'].unique()
```

```python
well_paid_data['Currency'] = well_paid_data['Currency'].str.replace('\t', ' ')
well_paid_data['Currency'] = well_paid_data['Currency'].apply(lambda x: x.split(' ')[0])

well_paid_data
```

```python
well_paid_data[well_paid_data['Currency']=='none']
```

```python
well_paid_data['Currency'] = well_paid_data['Currency'].str.replace('none', 'DKK')
```

```python
currency_rates = pd.read_csv('exchange_rates.csv')
currency_rates = currency_rates.iloc[:, 1:]

currency_rates = currency_rates[currency_rates['date'].str.contains('29/08/2024')]
currency_rates = currency_rates[['currency', 'value']]
currency_rates = currency_rates.rename(columns={'currency': 'Currency', 'value':'Value'})
currency_rates
```

```python
currency_rates_diff = pd.read_csv('currency-rates.csv')
currency_rates_diff = currency_rates_diff.iloc[:, 1:]
currency_rates_diff
```

```python
currency_rates_diff = currency_rates_diff[currency_rates_diff['Currency'].isin(['BYN', 'UZS', 'ZMW', 'TMT', 'TJS', 'IMP', 'VES', 'XDR', 'MRU', 'GGP'])]
```

```python
exchange_rate = pd.concat([currency_rates, currency_rates_diff], ignore_index=True)
exchange_rate
```

```python
well_paid_data
```

```python
well_paid_data_new = pd.merge(well_paid_data, exchange_rate, on='Currency', how='left')[['Country', 'Currency', 'CompTotal', 'Value']]
well_paid_data_new
```

```python
well_paid_data_new.isnull().sum()
```

```python
well_paid_data_new = well_paid_data_new[well_paid_data_new['CompTotal']>0]
```

```python
well_paid_data_new.dtypes
```

```python
well_paid_data_new['ConvertedCompYearlyUSD'] = round(well_paid_data_new['CompTotal'] / well_paid_data_new['Value'], 2)
well_paid_data_new
```

```python
# To avoid anomalies, it will be better to include countries more than 10 respondents

country_counts = well_paid_data_new['Country'].value_counts()

countries_to_keep = country_counts[country_counts > 10].index

well_paid_data_new = well_paid_data_new[well_paid_data_new['Country'].isin(countries_to_keep)]

well_paid_data_new
```

```python
well_paid_by_country = well_paid_data_new.groupby('Country')['ConvertedCompYearlyUSD'].mean().sort_values(ascending=False).reset_index()
well_paid_by_country_top = well_paid_by_country.head(11)
```

```python
well_paid_by_country_top
```

```python
import matplotlib.pyplot as plt

well_paid_by_country_top = well_paid_by_country_top.sort_values(by='ConvertedCompYearlyUSD', ignore_index=True)

plt.barh(well_paid_by_country_top['Country'], well_paid_by_country_top['ConvertedCompYearlyUSD'], color='#a4c2f4')
plt.xlabel('Countries')
plt.ylabel('Mean Salary')
plt.title('Mean Salary of Countries')
plt.show()
```

Here data should be normalized.

```python
import numpy as np
well_paid_by_country_top['NormConvertedCompYearlyUSD'] = np.log10(well_paid_by_country_top['ConvertedCompYearlyUSD'])
well_paid_by_country_top = well_paid_by_country_top.sort_values(by='ConvertedCompYearlyUSD', ascending=False, ignore_index=True)
well_paid_by_country_top
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

plt.scatter(well_paid_by_country_top['Country'], well_paid_by_country_top['NormConvertedCompYearlyUSD'], color='blue')

for i in range(well_paid_by_country_top.shape[0]):
    plt.text(x=i, y=well_paid_by_country_top['NormConvertedCompYearlyUSD'].iloc[i], s=well_paid_by_country_top['Country'].iloc[i], fontsize=9, ha='center')


plt.xticks(rotation=90)
plt.title('Scatter Plot to Identify Outliers')
plt.xlabel('Country')
plt.ylabel('Mean Salary')

plt.show()
```

```python
well_paid_by_country_top = well_paid_by_country_top[well_paid_by_country_top['Country'] != 'Afghanistan']
```

```python
plt.style.use('seaborn')

well_paid_by_country_top = well_paid_by_country_top.sort_values(by='ConvertedCompYearlyUSD', ascending=True, ignore_index=True)

plt.barh(well_paid_by_country_top['Country'], np.log10(well_paid_by_country_top['ConvertedCompYearlyUSD']), color='#a4c2f4')
plt.xlabel('Log10 of Converted Annual Compensation (USD)')
plt.ylabel('Country')
plt.title('Log10 of Annual Compensation by Country')
plt.show()
```

# How does coding experience affect the level of pay?

```python
pd.set_option('display.max_columns', len(df.columns))
df
```

```python
schema
```

```python
years_code_data = df[['YearsCode','YearsCodePro', 'Country', 'Currency', 'CompTotal', 'ConvertedCompYearly']]

years_code_data = years_code_data.dropna(how='all')
```

```python
years_code_data = years_code_data.dropna(subset=['YearsCode','YearsCodePro'], how='all')
years_code_data = years_code_data.dropna(subset=['Currency','CompTotal', 'ConvertedCompYearly'], how='all')
years_code_data = years_code_data.dropna(subset=['CompTotal', 'ConvertedCompYearly'], how='all')
years_code_data
```

```python
years_code_data['YearsCodePro'] = years_code_data['YearsCodePro'].fillna(years_code_data['YearsCode'])
years_code_data['YearsCode'] = years_code_data['YearsCode'].fillna(years_code_data['YearsCodePro'])

years_code_data.isnull().sum()
```

```python
years_code_data['Currency'] = years_code_data['Currency'].str.replace('\t', ' ')
years_code_data['Currency'] = years_code_data['Currency'].apply(lambda x: x.split(' ')[0])
years_code_data
```

```python
years_code_data = pd.merge(years_code_data, exchange_rate, on='Currency', how='left')[['YearsCodePro', 'Country', 'Currency', 'CompTotal', 'Value']]
years_code_data
```

```python
years_code_data.isnull().sum()
```

```python
years_code_data[years_code_data.isna().any(axis=1)]
```

```python
years_code_data['Currency'] = years_code_data['Currency'].str.replace('none', 'DKK')
```

```python
import numpy as np
years_code_data['ConvertedCompYearlyUSD'] = round(years_code_data['CompTotal'] / years_code_data['Value'], 2)
years_code_data = years_code_data[years_code_data['Country'] != 'Afghanistan']
years_code_data = years_code_data[years_code_data['ConvertedCompYearlyUSD']>0]
years_code_data['NormConvertedCompYearlyUSD'] = np.log10(years_code_data['ConvertedCompYearlyUSD'])
```

```python
years_code_data
```

```python
years_code_data_res = years_code_data[['YearsCodePro', 'ConvertedCompYearlyUSD', 'NormConvertedCompYearlyUSD']]
years_code_data_res
```

```python
years_code_data_res['YearsCodePro'].unique()
```

```python
years_code_data_res['YearsCodePro'] = years_code_data_res['YearsCodePro'].replace('Less than 1 year', 0.5)
years_code_data_res['YearsCodePro'] = years_code_data_res['YearsCodePro'].replace('More than 50 years', 51)
years_code_data_res['YearsCodePro'] = years_code_data_res['YearsCodePro'].astype(float)
years_code_data_res = years_code_data_res[['YearsCodePro', 'ConvertedCompYearlyUSD', 'NormConvertedCompYearlyUSD']]
years_code_data_res
```

```python
years_code_data_res['YearsCodePro'].unique()
```

```python
bins = [0.0, 5.0, 10.0, 15.0, 20.0, 25.0, 30.0, 35.0, 40.0, 50.0, np.inf]
labels = ['0-5', '6-10', '11-15', '16-20', '21-25', '26-30', '31-35', '36-40', '41-50', '51+']
years_code_data_res['ExperienceRange'] = pd.cut(years_code_data_res['YearsCodePro'], bins=bins, labels=labels)
years_code_data_res
```

```python
mean_salary_by_experience = years_code_data_res.groupby('ExperienceRange')['ConvertedCompYearlyUSD'].mean().reset_index()
mean_salary_by_experience['NormConvertedCompYearlyUSD'] = np.log10(mean_salary_by_experience['ConvertedCompYearlyUSD'])
mean_salary_by_experience
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


plt.figure(figsize=(12, 6))
plt.scatter(x='ExperienceRange', y='NormConvertedCompYearlyUSD', data=mean_salary_by_experience, color='blue')

#for i in range(well_paid_by_country_top.shape[0]):
 #   plt.text(x=i, y=well_paid_by_country_top['NormConvertedCompYearlyUSD'].iloc[i], s=well_paid_by_country_top['Country'].iloc[i], fontsize=9, ha='center')

plt.title('Log10 of Average Salary Distribution by Experience Range')
plt.xlabel('Years of Professional Coding Experience (Grouped)')
plt.ylabel('Log10 of Converted Compensation Yearly (USD)')
plt.show()
```

# What's the most popular method of learning to code?

```python
learning_code_data = df[['LearnCode']]
learning_code_data = learning_code_data.dropna()
learning_code_data 
```

```python
learning_code_data['LearnCode'].unique()
```

```python
pattern = r'^([^;|:]*)'
learning_code_data['Extracted Statements'] = learning_code_data['LearnCode'].str.extract(pattern)
learning_code_data
```

```python
learning_code_data['Extracted Statements'].unique()
```

```python
learning_code_data = learning_code_data.groupby('Extracted Statements').size().reset_index(name='Count')
learning_code_data 
```

```python
plt.style.use('seaborn')

plt.figure(figsize=(12, 6))

colors = ['#d9ead3', '#d0e0e3', '#c9daf8', '#b6d7a8', '#a2c4c9', '#a4c2f4', '#93c47d', '#76a5af', '#6d9eeb']

plt.pie(learning_code_data['Count'], labels=learning_code_data['Extracted Statements'], startangle=90, autopct='%1.1f%%',wedgeprops={'edgecolor': 'black'}, colors=colors)

plt.title('Methods of learning code')
plt.tight_layout()
plt.show()
```

# Are you more likely to get a job as a developer if you have a master's degree?

```python
schema
```

```python
emp_edu_data = df[['Employment', 'EdLevel', 'DevType']]
emp_edu_data
```

```python
emp_edu_data.isnull().sum()
```

```python
emp_edu_data['Employment'].unique()
```

```python
emp_edu_data = emp_edu_data[emp_edu_data['Employment'].str.contains('Employed|self-employed', case=False)]
emp_edu_data = emp_edu_data[~emp_edu_data['Employment'].str.contains('Not employed', case=False)]
emp_edu_data
```

```python
emp_edu_data['Employment'].unique()
```

```python
emp_edu_data['Employment'] = emp_edu_data['Employment'].apply(
    lambda x: 'Employed/Self-employed' if 'Employed' in str(x) or 'self-employed' in str(x) else x
)
emp_edu_data
```

```python
emp_edu_data['Employment'].unique()
```

```python
emp_edu_data['EdLevel'].unique()
```

```python
emp_edu_data = emp_edu_data.dropna(subset=['EdLevel'])
emp_edu_data
```

```python
emp_edu_data = emp_edu_data.dropna(subset=['DevType'])
emp_edu_data 
```

```python
emp_edu_data['DevType'] = emp_edu_data['DevType'].apply(lambda x: 'Developer' if 'Developer' in str(x) else x)
emp_edu_data = emp_edu_data.dropna(subset=['DevType'])
emp_edu_data = emp_edu_data[emp_edu_data['DevType'] == 'Developer']
emp_edu_data
```

```python
emp_edu_data['EdLevel'].unique()
```

```python
emp_edu_data['EdLevel'].isnull().sum()
```

```python
emp_edu_data = emp_edu_data.groupby('EdLevel').size().reset_index(name='Count')
emp_edu_data
```

```python
plt.style.use('seaborn')

plt.figure(figsize=(12, 6))

colors = ['#d9ead3', '#d0e0e3', '#c9daf8', '#b6d7a8', '#a2c4c9', '#93c47d', '#76a5af', '#6d9eeb']

plt.pie(emp_edu_data['Count'], labels=emp_edu_data['EdLevel'], startangle=90, autopct='%1.1f%%',wedgeprops={'edgecolor': 'black'}, colors=colors)

plt.title('Number of Employeed People as Developers with Different Educational Degrees')
plt.tight_layout()
plt.show()
```
