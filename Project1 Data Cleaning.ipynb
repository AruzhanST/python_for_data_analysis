# Q1: Can you make the salary column into integers?

```python
df.head(10)
```

```python
df['Salary Estimate'].unique()
```

```python
df['Salary Estimate dup'] = df['Salary Estimate']
```

```python
df['Salary Estimate dup'] = df['Salary Estimate dup'].str.replace('Employer Provided Salary:', '')
df['Salary Estimate dup'] = df['Salary Estimate dup'].str.replace(' \(Glassdoor est.\)', '')
df.head(15)
```

```python
df['Salary Estimate dup'].unique()
```

```python
import pandas as pd

def parsing_salary(salary_str):
    if 'Per Hour' in salary_str:
        salary_str = salary_str.replace('Per Hour', '').replace('$', '').strip()
        
        parts = salary_str.split(' - ')
        if len(parts) != 2:
            min_salary = max_salary = float(salary_str)*40*52
            return pd.Series([min_salary, max_salary], index=['Min Salary', 'Max Salary'])
        try:
            min_salary, max_salary = map(float, parts)
            min_salary = min_salary * 40 * 52
            max_salary = max_salary * 40 * 52
            return pd.Series([min_salary, max_salary], index=['Min Salary', 'Max Salary'])
        except ValueError:
            return pd.Series([None, None], index=['Min Salary', 'Max Salary'])
        
        
    elif 'K' in salary_str:
        salary_str = salary_str.replace('$', '').replace('K', '').strip()
        parts = salary_str.split(' - ')
        if len(parts)==1:
            min_salary = max_salary = int(salary_str)*1000
            return pd.Series([min_salary, max_salary], index=['Min Salary', 'Max Salary'])
        if len(parts) != 2:
            return pd.Series([None, None], index=['Min Salary', 'Max Salary'])
        try:
            min_salary, max_salary = map(float, parts)
            min_salary = min_salary*1000
            max_salary = max_salary*1000
            return pd.Series([min_salary, max_salary], index=['Min Salary', 'Max Salary'])
        except ValueError:
            return pd.Series([None, None], index=['Min Salary', 'Max Salary'])
    else:
        return pd.Series([None, None], index=['Min Salary', 'Max Salary'])


df[['Min Salary', 'Max Salary']] = df['Salary Estimate dup'].apply(parsing_salary)
```

```python
df.loc[:, 'Min Salary'] = df['Min Salary'].astype(int)
df.loc[:, 'Max Salary'] = df['Max Salary'].astype(int)
```

```python
df.drop(columns='Salary Estimate dup', inplace=True)
df.head(5)
```

# Q2: What information can you extract out of job descriptions?

```python
df['Job Description'].nunique()
```

```python
df['Job Description'].unique()
```

```python
import pandas as pd
import re

keywords = ['we', 'our', 'company', 'job', 'agency', 'team', 'work', 'seeking', 'working', 'position', 'data', 'medicine', 'duties', 'responsibilities', 'competitive pay']

# Function to clean and extract statements
def extract_statements(text, keywords, num_statements=3):
    # Clean text
    text = re.sub(r'\n+', ' ', text)  # Replace newlines with spaces
    text = re.sub(r'\s+', ' ', text)  # Replace multiple spaces with a single space
    text = re.sub(r'\\u[0-9A-Fa-f]{4}', '', text)
    
    text = text.lower() 
    
    statements = re.split(r'\.\s*|\!\s*|\?\s*', text) #so, 'statements' hold a list
    
    relevant_statements = [stmt for stmt in statements if any(keyword in stmt for keyword in keywords)] 
    
    return ';'.join(relevant_statements[:num_statements])


df['Extracted Statements'] = df['Job Description'].apply(lambda x: extract_statements(x, keywords))
```

# Q3: How can you remove the numbers from the company name?

```python
df['Company Name'].unique()
```

```python
pattern = r'\n\d\.\d'

try:
    df['Company Name'] = df['Company Name'].apply(lambda x: re.sub(pattern ,'', x))
except Exception as e:
    print(f'error is {e}')

df.head(5)
```

# Q4: How can you create some new features? (e.g. state column from the location column)

```python
df['Location'].unique()
```

```python
try:
    df['State'] = df['Location'].apply(lambda x: re.split(',', x)[-1])
    df['Location'] = df['Location'].replace({'-1': 'Missing'})
    df['State'] = df['State'].replace({
        'California': 'CA',
        'United States': 'USA',
        'Maryland': 'MD',
        'Illinois': 'IL',
        'New York State': 'NY',
        'Point Loma': 'CA',
        'Manhattan': 'NY',
        '-1': 'Missing'
    })
except Exception as e:
    print(f'{e}')

df.head(5)
```
