```python
# AS IT IS SEEN, WE CAN USE 'INDEX' COLUMN AS INDEX
df = pd.read_csv('data.csv', index_col='Index')
df
```

```python
# First, let's work with NaN values in 'Age' column

df['Age'].fillna(0, inplace=True)
```

```python
df
```

```python
# LET'S CHECK DATATYPES OF COLUMNS

df.dtypes
```

```python
# LET'S WORK WITH 'SALARY' COLUMN
df[['Lowest salary', 'Highest salary']] = df['Salary'].str.split('-', expand=True)
df
```

```python
# LET'S DROP 'SALARY' COLUMN
df.drop(columns='Salary', inplace=True) 
df
```

```python
# 'Lowest salary' and 'Highest salary' columns should be modified
df['Lowest salary'] = df['Lowest salary'].str.strip('$k')
df['Highest salary'] = df['Highest salary'].str.strip('$k')
```

```python
df['Lowest salary'] = df['Lowest salary'] + '000'
df['Highest salary'] = df['Highest salary'] + '000'
```

```python
df
```

```python
# Let's change datatypes of 'Lowest salary' and 'Highest salary' columns to INT
df['Lowest salary'] = df['Lowest salary'].astype(int)
df['Highest salary'] = df['Highest salary'].astype(int)
```

```python
df.dtypes
```

```python
# Range of 'Rating' column probably should be between 0 and 10. Let's change it to correct form.
df['Rating'].unique()
```

```python
# So '-1' and 'nan' values should be changed.
df['Rating'].replace([-1, np.nan], 0, inplace=True)
```

```python
df
```

```python
# In 'Established' column, wrong values should be modified as this column only contains years.

df['Established'] = df['Established'].replace(-1, '')
```

```python
df
```

```python
# In 'Easy Apply' column, let's change 'TRUE' to 'YES' and '-1' to 'NO'
df['Easy Apply'] = df['Easy Apply'].replace('TRUE', 'YES')
df['Easy Apply'] = df['Easy Apply'].replace('-1', 'NO')
```

```python
df
```

```python
# Let's consider name of city/country and their abbreviations. Regular expressions can be applied.
pattern = r'(\w+ \w+|\w+)[ ,](\w+)' 
df[['Country/City', 'State']] = df['Location'].str.extract(pattern)
df.drop(columns='Location', inplace=True) 
```

```python
# Finally, this is cleaned dataframe
df
```
